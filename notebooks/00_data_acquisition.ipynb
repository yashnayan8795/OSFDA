# Notebook 00: Data Acquisition
This notebook handles the initial download of the ASRS dataset from Hugging Face and performs a basic inventory of the columns.

```python
import pandas as pd
import yaml
from datasets import load_dataset
from pathlib import Path
import os

# Change to the 'rebuild' directory if not already there
# os.chdir('..') # Adjust based on where you run the notebook

with open('configs/main_config.yaml', 'r') as f:
    config = yaml.safe_load(f)

print(f"Project: {config['project']['name']}")
```

## 1. Load Dataset

```python
print(f"Loading dataset: {config['data']['hf_dataset']}")
dataset = load_dataset(config['data']['hf_dataset'])
df = dataset['train'].to_pandas()

print(f"Dataset loaded. Shape: {df.shape}")
```

## 2. Column Inventory

```python
inventory = []
for col in df.columns:
    inventory.append({
        'column': col,
        'dtype': str(df[col].dtype),
        'missing_pct': df[col].isna().mean() * 100,
        'unique_count': df[col].nunique(),
        'samples': df[col].dropna().unique()[:5].tolist()
    })

inventory_df = pd.DataFrame(inventory)
display(inventory_df.sort_values('missing_pct', ascending=False).head(20))

inventory_df.to_csv('data/raw/column_inventory.csv', index=False)
```

## 3. Save Raw Data

```python
save_path = config['paths']['raw_data']
df.to_parquet(save_path, index=False)
print(f"Raw data saved to {save_path}")
```

Download the ASRS dataset, inspect columns, and save raw parquet.

```python
import os
from pathlib import Path

# Find project root (contains pyproject.toml)
root = Path.cwd()
while not (root / "pyproject.toml").exists():
    root = root.parent
os.chdir(root)
print(f"Working directory: {root}")
```

```python
from src.utils.config import load_main_config, resolve_path
from src.data.loader import load_raw_data, column_inventory

config = load_main_config()
print(f"Project: {config['project']['name']}")
```

## 1. Load Dataset

```python
df = load_raw_data(config)
print(f"Dataset loaded. Shape: {df.shape}")
```

## 2. Column Inventory

```python
inv = column_inventory(df)
display(inv.head(20))

inv_path = resolve_path('data/raw/column_inventory.csv')
inv.to_csv(inv_path, index=False)
print(f'Saved to {inv_path}')
```

## 3. Save Raw Data

```python
save_path = resolve_path(config['paths']['raw_data'])
df.to_parquet(save_path, index=False)
print(f'Raw data saved to {save_path}')
```
